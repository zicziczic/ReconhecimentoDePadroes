```python
import pandas as pd
import os
import glob
import shutil
```

```python
benignant_path = "./benignant_skin_scaled.csv"
malignant_path = "./malignant_skin_scaled.csv"
```

```python
benignant_df = pd.read_csv(benignant_path, index_col=0)
benignant_df = benignant_df.reset_index(drop=True)
malignant_df = pd.read_csv(malignant_path, index_col=0)
malignant_df = malignant_df.reset_index(drop=True)
```

```python
benignant_ids = benignant_df["isic_id"].to_list()
benignant_ids = [name + ".jpg" for name in benignant_ids]
malignant_ids = malignant_df["isic_id"].to_list()
malignant_ids = [name + ".jpg" for name in malignant_ids]
```

```python
images_path = "./data/data/"
gen_path = "./data/"
```

```python
all_images = glob.glob(os.path.join(images_path, "*.jpg"))
len(all_images)
```

```python
os.makedirs(gen_path+"benignant_images", exist_ok=True)
os.makedirs(gen_path+"malignant_images", exist_ok=True)
```

```python
benignant_folder = gen_path+"benignant_images/"
malignant_folder = gen_path+"malignant_images/"


for file_name in benignant_ids:
    src_path = os.path.join(images_path, file_name)
    dst_path = os.path.join(benignant_folder, file_name)

    # Only copy files (skip subfolders)
    if os.path.isfile(src_path):
        shutil.copy(src_path, dst_path)
        
for file_name in malignant_ids:
    src_path = os.path.join(images_path, file_name)
    dst_path = os.path.join(malignant_folder, file_name)

    # Only copy files (skip subfolders)
    if os.path.isfile(src_path):
        shutil.copy(src_path, dst_path)
```
