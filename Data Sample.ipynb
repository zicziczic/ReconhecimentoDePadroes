```python
c = df["patient_id"].value_counts()
c = c[c>10]
df_pelomenos10 = df[df["patient_id"].isin(c.index)]
df_pelomenos10 = df_pelomenos10.iloc[:, :-12]
#df_pelomenos10 = df_pelomenos10.dropna()
```

```python
df_pelomenos10
```

```python
pintas_malignas = df_pelomenos10[df_pelomenos10["target"] == 1]
pessoas_pmalignas = pintas_malignas["patient_id"].unique()
n_pessmalignas = len(pessoas_pmalignas)
df_pintas_pmalignas = df_pelomenos10[df_pelomenos10["patient_id"].isin(pessoas_pmalignas)]
print(len(pintas_malignas), len(pessoas_pmalignas), len(df_pintas_pmalignas))
```

```python
# pibegnigna_pmaligna representam 5 imagens de pintas benignas de pessoas com pintas malignas
pibegnigna_pmaligna = df_pintas_pmalignas[df_pintas_pmalignas["target"]==0].groupby("patient_id").sample(5)
#juntar pibegnigna_pmaligna com pintas_malignas (calculado na celula acima) -- vai ter que ter 386+1285 linhas
```

```python
pessoas_psobenignas = list(set(df_pelomenos10["patient_id"]) - set(pessoas_pmalignas))
print("Total de pessoas com apenas pintas benignas: %d" % len(pessoas_psobenignas))
np.random.seed(70)
#pessoas_psobenignas_amostra = np.random.choice(pessoas_psobenignas, n_pessmalignas)
np.random.shuffle(pessoas_psobenignas)
pessoas_psobenignas_amostra = pessoas_psobenignas[:n_pessmalignas]
```

```python
len(pessoas_pmalignas), len(pessoas_psobenignas_amostra)
```

```python
df_pintas_pbenignas = df_pelomenos10[df_pelomenos10["patient_id"].isin(pessoas_psobenignas_amostra)]
df_pintas_pbenignas = df_pintas_pbenignas.groupby("patient_id").sample(5)

pibegnigna_pmaligna["skin_type"] = "malignant"
pintas_malignas["skin_type"] = "malignant"
```

```python
df_pintas_pbenignas["skin_type"] = "benignant"
```

```python
df_pintas_pbenignas.to_csv("./Benignant_Skin_Type.csv")
```
